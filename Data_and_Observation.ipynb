```python
import numpy as np
import pandas as pd
STUDENT_NAME = "Aisy Cordero"
STUDENT_ID = 250354

np.random.seed(STUDENT_ID)

# Creating a DataFrame with Education Levels and random Salaries
df = pd.DataFrame({
    'EdLevel': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master'],
    'ConvertedComp': np.random.randint(40000, 150000, 5)
})

# Calculate mean, median, and standard deviation for each 'EdLevel'
summary = df.groupby('EdLevel')['ConvertedComp'].agg(['mean', 'median', 'std']).reset_index()

print(f"{STUDENT_NAME}'s Grouped Summary Statistics:\n", summary)
```

```python
import numpy as np
import pandas as pd
STUDENT_NAME = "Aisy Cordero"
STUDENT_ID = 250354

np.random.seed(STUDENT_ID)

# Generate salaries based on your student ID seed
df = pd.DataFrame({'ConvertedComp': np.random.randint(40000, 150000, 5)})

threshold = 110000

# Create the 'HighPay' column using np.where()
df['HighPay'] = np.where(df['ConvertedComp'] > threshold, 'Yes', 'No')

print(f"{STUDENT_NAME}'s Salary Classification:\n", df)
```
