```python
import numpy as np

try:
    # Array 'a' has shape (3,) - a 1D array
    a = np.array([1, 2, 3])
    
    # Array 'b' has shape (3, 1) - a 2D column vector
    b = np.array([[1], [2], [3]])
    
    # Attempting addition
    result = a + b
    
    print("Result of addition:")
    print(result)

except ValueError as e:
    print(f"Broadcasting error: {e}")
```

```python
import numpy as np

# Generates a 3x3 identity matrix (1s on the diagonal, 0s elsewhere)
identity = np.eye(3)

# A standard 3x3 matrix
matrix = np.array([[2, 3, 4], [1, 0, 6], [7, 8, 9]])

# Matrix addition
result = matrix + identity

print("Matrix + Identity:\n", result)
```

```python
import numpy as np

# An array containing a Not-a-Number (NaN) and an Infinity (inf)
data = np.array([1, 2, np.nan, 4, np.inf])

# np.isfinite returns False for NaN and Inf; True for normal numbers
cleaned = data[np.isfinite(data)]

print("Cleaned Data:", cleaned)
```

```python
import pandas as pd

# Creating a DataFrame with missing values (None)
df = pd.DataFrame({
    "Name": ["Kurt", "Marck", "Daddy Bao", "Joecel"],
    "Score": [85, None, 92, 88]
})

# Removing any row that contains at least one missing value
df_cleaned = df.dropna()

print("Cleaned DataFrame:\n", df_cleaned)
```

```python
import pandas as pd
import numpy as np

# Creating a simple numeric DataFrame
df = pd.DataFrame({
    "X": [1, 2, 3],
    "Y": [4, 5, 6]
})

# Applying NumPy's log and square root functions directly to columns
df["Log_X"] = np.log(df["X"])
df["Sqrt_Y"] = np.sqrt(df["Y"])

print("Transformed DataFrame:\n", df)
```
